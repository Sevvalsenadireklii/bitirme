```python
def gunlukArtis(data):
    liste = []
    for i in range(len(data)):
        if i ==0:
            liste.append(data[0])
        else:
            liste.append(data[i]-data[i-1])
            
    return liste
```

```python
def country_plot(gecenZaman, ulkeVaka, 
                 gunlukVaka, gunlukOlum, 
                 gunlukKurtulma, country):
    
    plt.figure(figsize=(10,8))
    plt.bar(gecenZaman, ulkeVaka,width=1,
        color="#f9af2b",edgecolor="#f05131")
    plt.title("{} ' nin Vaka Sayısı".format(country))
    plt.xlabel(" Geçen Zaman")
    plt.ylabel("Vaka Sayısı")

    
    plt.figure(figsize=(10,8))
    plt.bar(gecenZaman, gunlukVaka,width=1,
        color="#f9af2b",edgecolor="#f05131")
    plt.title("{} ' nin Günlük Vaka Sayısı".format(country))
    plt.xlabel("Geçen Zaman")
    plt.ylabel("Günlük Vaka Sayısı")
 
        
    plt.figure(figsize=(10,8))
    plt.bar(gecenZaman, gunlukOlum,width=1,
        color="#f9af2b",edgecolor="#f05131")
    plt.title("{} ' nin Günlük Ölüm Sayısı".format(country))
    plt.xlabel("Geçen Zaman")
    plt.ylabel("Günlük Ölüm Sayısı")
    
    
    plt.figure(figsize=(10,8))
    plt.bar(gecenZaman, gunlukKurtulma,width=1,
        color="#f9af2b",edgecolor="#f05131")
    plt.title("{} ' nin Günlük Kurtulma Sayısı".format(country))
    plt.xlabel("Geçen Zaman")
    plt.ylabel("Günlük Kurtulma Sayısı")
```

```python
country_plot(gecenZaman, chinaVaka, china_gunlukArtis, china_olumArtis, china_kurtulmaArtis, "China")
```
